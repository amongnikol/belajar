# Data Cleaning Exercise #1 (Titanic)

```python
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/stefkwan-h8/dataset/main/titanic/train.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.info()
```

## identifikasi missing values

saya tidak melihat adanya invalid value atau inconsistent pada dataset

```python
df.isnull().sum()
```

## membuat rata rata nilai pada kolom Age lalu mengisi nilai yang kosong dengan nilai rata rata tersebut

```python
# rata-rata age
age_mean = int(df['Age'].mean())

df['Age'].fillna(age_mean, inplace=True)
df.info()
```

```python
# mengubah tipe data kolom Age menjadi int agar tidak ada nilai desimal
df['Age'] = df['Age'].astype(int)
df.info()
```

## menghapus kolom Cabin karena terlalu banyak jumlah baris yang kosong

```python
df.drop('Cabin', axis=1, inplace=True)

df.info()
```

## mengisi nilai yang kosong pada kolom Embarked dengan nilai terbanyak pada kolom tersebut

```python
embarked_modus = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(embarked_modus)
df.info()
```

saya tidak melihat adanya data yang duplikat

```python
df[df.duplicated()]
```

## mengubah tipe data pada kolom kolom tertentu untuk mengurangi kapasitas memori

```python
df['Pclass'].value_counts()
```

```python
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'] = df['Sex'].astype('category')
```

```python
df['PassengerId'] = df['PassengerId'].astype('int32')
```

```python
df.info()
```

## melihat outlier pada kolom Fare

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
# Filter
df_filtered = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]

# Plot boxplot

df_filtered['Fare'].plot(kind='box')
```

## normalisasi data pada kolom Fare

```python
from sklearn.preprocessing import MinMaxScaler
# scaler
scaler = MinMaxScaler()
# Normalisasi kolom 'Fare'
df[['Fare']] = scaler.fit_transform(df[['Fare']])
# Tampilkan data yang sudah di normalisasi
print(df[['Fare']].head())
```

# Data Cleaning Exercise #2 (open-ended)

```python
import numpy as np
import pandas as pd


df_pelanggan = pd.read_csv('https://raw.githubusercontent.com/amongnikol/belajar/refs/heads/main/dataset/Customers.csv')
df_pelanggan.head(10)
```

## handling missing values

```python
df_pelanggan.info()
```

```python
# cek kolom Spending Score (1-100) apakah ada nilai yang tidak sesuai dari ketentuan
cek_spending = df_pelanggan['Spending Score (1-100)'] > 100
cek_spending2 = df_pelanggan['Spending Score (1-100)'] < 1
print(cek_spending)
print(cek_spending2)
```

```python
df_pelanggan['Work Experience'].value_counts()
```

```python
# menghapus kolom work experience dikarenakan terlalu banyak data yang tidak sesuai yaitu pada nilai 0
df_pelanggan.drop('Work Experience', axis=1, inplace=True)

df_pelanggan.info()
```

```python
profession_modus = df_pelanggan['Profession'].mode()[0]
df_pelanggan['Profession'] = df_pelanggan['Profession'].fillna(profession_modus)
df_pelanggan.info()
```

## handling duplicate

```python
# saya tidak melihat adanya data duplikat
df_pelanggan[df_pelanggan.duplicated()]
```

## menyesuaikan tipe data pada kolom

```python
df_pelanggan['Gender'].value_counts()
```

```python
df_pelanggan['Gender'] = df_pelanggan['Gender'].astype('category')
df_pelanggan.info()
```

```python
df_pelanggan['Profession'].value_counts()
```

```python
df_pelanggan['Profession'] = df_pelanggan['Profession'].astype('category')
df_pelanggan.info()
```

## handling outlier

```python
# saya tidak melihat adanya outlier
Q1 = df_pelanggan['Annual Income ($)'].quantile(0.25)
Q3 = df_pelanggan['Annual Income ($)'].quantile(0.75)
IQR = Q3 - Q1
# Filter
df_filtered = df_pelanggan[(df_pelanggan['Annual Income ($)'] >= Q1 - 1.5 * IQR) & (df_pelanggan['Annual Income ($)'] <= Q3 + 1.5 * IQR)]

# Plot boxplot

df_filtered['Annual Income ($)'].plot(kind='box')
```

## normalisasi data pada kolom Annual income ($)

```python
from sklearn.preprocessing import MinMaxScaler
# scaler
scaler = MinMaxScaler()
# Normalisasi kolom 'Annual Income ($)'
df_pelanggan[['Annual Income ($)']] = scaler.fit_transform(df_pelanggan[['Annual Income ($)']])
# Tampilkan data yang sudah di normalisasi
print(df_pelanggan[['Annual Income ($)']].head())
```
